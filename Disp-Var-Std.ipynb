## Generating a Lot of Numbers

```python
num_samples=10000
```

```python
uniform_vals=np.random.uniform(low=0.0, high=10.0, size=num_samples)
```

```python
sns.distplot(uniform_vals, bins=20, kde=False)
plt.ylabel('Frequency')
plt.xlabel('Value')
plt.title("Uniform Values")
sns.despine(offset=10, trim=True)
plt.show()
```

```python
normal_vals=np.random.normal(loc=5.0, scale=10.0, size=num_samples)
```

```python
sns.distplot(normal_vals, bins=20, kde=False)
plt.ylabel('Frequency')
plt.xlabel('Value')
plt.title("Uniform Values")
sns.despine(offset=10, trim=True)
plt.show()
```

```python
print("Uniform values mean:",np.mean(uniform_vals))
print("Normal values mean:",np.mean(normal_vals))
```

```python
print("Uniform values var:",np.var(uniform_vals))
print("Normal values var:",np.var(normal_vals))
```

```python
print("Uniform values std:",np.std(uniform_vals))
print("Normal values std:",np.std(normal_vals))
```

```python
normal_vals=np.random.normal(loc=5.0, scale=10.0, size=num_samples)
nv_mean=np.mean(normal_vals)
nv_std=np.std(normal_vals)
```

```python
sns.distplot(normal_vals, bins=20, kde=False)
plt.xlim(-60,60)

plt.ylabel("Frequency")
plt.xlabel('Value')
plt.title("Normal Values")
sns.despine(offset=10, trim=True)

x_c, y_c = ([nv_mean, nv_mean], [0,1500])
plt.plot(x_c, y_c, color='green', linewidth=2)

x_c, y_c = ([nv_mean + nv_std, nv_mean + nv_std], [0,1500])
plt.plot(x_c, y_c, color="red", linewidth=2)

x_c, y_c = ([nv_mean - nv_std, nv_mean - nv_std], [0,1500])
plt.plot(x_c, y_c, color='red', linewidth=2)

plt.show()
```

```python
normal_vals=np.random.normal(loc=5.0, scale=15.0, size=num_samples)
nv_mean=np.mean(normal_vals)
nv_std=np.std(normal_vals)
```

```python
sns.distplot(normal_vals, bins=20, kde=False)
plt.xlim(-60,60)

plt.ylabel("Frequency")
plt.xlabel('Value')
plt.title("Normal Values")
sns.despine(offset=10, trim=True)

x_c, y_c = ([nv_mean, nv_mean], [0,1500])
plt.plot(x_c, y_c, color='green', linewidth=2)

x_c, y_c = ([nv_mean + nv_std, nv_mean + nv_std], [0,1500])
plt.plot(x_c, y_c, color="red", linewidth=2)

x_c, y_c = ([nv_mean - nv_std, nv_mean - nv_std], [0,1500])
plt.plot(x_c, y_c, color='red', linewidth=2)

plt.show()
```

```python
normal_vals = np.random.normal(loc=5.0, scale=10.0, size=num_samples)
normal_vals2 = np.random.normal(loc=5.0, scale=15.0, size=num_samples)

def plot_dist(vals, label):
    nv_mean = np.mean(vals)
    nv_std = np.std(vals)
    
    sns.distplot(vals, bins=20, kde=False)
    
    plt.xlim(-60,60)
    
    plt.ylabel('Frequency')
    plt.xlabel('Value')
    plt.title(label)
    sns.despine(offset=10, trim=True)
    
    x_c, y_c = ([nv_mean, nv_mean], [0,1500])
    plt.plot(x_c, y_c, color='green', linewidth=1)
    
    x_c, y_c = ([nv_mean + nv_std, nv_mean + nv_std], [0,1500])
    plt.plot(x_c, y_c, color='red', linewidth=1)
    
    x_c, y_c = ([nv_mean - nv_std, nv_mean - nv_std], [0,1500])
    plt.plot(x_c, y_c, color='red', linewidth=1)
    
    plt.show()
```

```python
plot_dist(normal_vals, "Less Variant")
```

```python
plot_dist(normal_vals2, "More Variant")
```

```python
med_inc = [x[0] for x in houses.data]
mean = np.mean(med_inc)
std = np.std(med_inc)

sns.distplot(med_inc, bins=20, kde=False)

plt.ylabel('Frequency')
plt.xlabel('Median Income')
plt.title('Histogram of Median Incomes')
sns.despine(offset=10, trim=True)

x_c, y_c = ([mean, mean], [0,4000])
plt.plot(x_c, y_c, color='green', linewidth=2)

x_c, y_c = ([mean+std, mean+std], [0,4000])
plt.plot(x_c, y_c, color='red', linewidth=2)

x_c, y_c = ([mean-std, mean-std], [0,4000])
plt.plot(x_c, y_c, color='red', linewidth=2)

plt.show()
```
